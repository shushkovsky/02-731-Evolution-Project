```python
import numpy as np
import matplotlib as plt
import seaborn as sns
import pandas as pd
```

```python
with open(path3, 'r') as file:
    lines = file.readlines()
    print(lines[19])
```

```python
data = pd.read_csv(path3, skiprows=18, sep='\t', header=None)
columns = ["Index", "Language", "Language name in DB", "ISO code", "Highest Level Clasification", "Population", "Geo Region", "Latitude", "Longitude"]
```

```python
metadata = data.iloc[:, 0:9]
metadata.columns = columns
metadata
```

```python
cols_p = ["ph"+str(i+1) for i in range(728)]
```

```python
phonemes = data.iloc[:, 9:]
phonemes.columns = cols_p
phonemes
```

```python
df = pd.concat([metadata, phonemes], axis = 1)
df
```

```python
sns.clustermap(
    phonemes,
    figsize=(7, 5),
    row_cluster=False,
    dendrogram_ratio=(.1, .2),
    cbar_pos=(0, .2, .03, .4)
)
```

```python
sns.clustermap(phonemes, cmap="mako", vmin=0, vmax=10)
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(16,7))
sns.scatterplot(x='Latitude',y='Longitude',hue='Geo Region',data=df,
               legend="full")
plt.show()
```
